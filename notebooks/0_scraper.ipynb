```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import regex
import requests
import json
import os
from google.colab import drive
from datetime import datetime, timezone

drive.mount('/content/drive')
root = '/content/drive/MyDrive/Desys_Group/data/'
```

```python
def pull_data_from_opensea(N_iter=100, event_type_="successful", occurred_after=datetime(2022, 1, 1, tzinfo=timezone.utc).timestamp(), next_cursor_=""):
  url = "https://api.opensea.io/api/v1/events"

  if event_type_ == "all":
    querystring = {"only_opensea": "true", "occurred_after": occurred_after}
  else:
    querystring = {"only_opensea": "true", "event_type": event_type_, "occurred_after": occurred_after}

  if next_cursor_:
    querystring["cursor"] = next_cursor_

  headers = {
  	"Accept": "application/json",
  	"X-API-Key": "x-api-key"
  }

  event_list = []
  N_iter = N_iter

  for i in range(N_iter):
    response = requests.request("GET", url, headers=headers, params=querystring)
    next_cursor = response.json()["next"]
    querystring["cursor"] = next_cursor
    event_list += response.json()["asset_events"]
    last_created_date = event_list[-1]["created_date"]
    print("Round %d: next cursor is at %s" % (i, next_cursor), "(%s)" % (last_created_date))

  with open(os.path.join(root, 'json_data_%s_%s.json' % (event_type_, next_cursor)), 'w') as outfile:
    json_string = json.dumps(event_list)
    json.dump(json_string, outfile)
    print("Save file at %s" % os.path.join(root, 'json_data_%s_%s.json' % (event_type_, next_cursor)))
```
