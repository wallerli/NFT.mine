```python
# combine json array
def combine_json_array(ja1, ja2):
    return ja1 + ja2

# parse json data to DataFrame
def fill_dict(to_fill, json_data, key_path):
    for k, v in json_data.items():
        new_key = key_path + '/' + k
        if isinstance(v, dict):
            fill_dict(to_fill, v, new_key)
        elif v:
            to_fill[new_key] = v

def parse_json_string(json_string):
    new_data = []
    for d in json.loads(json_string):
        new_dict = {}
        fill_dict(new_dict, d, 'event')
        new_data.append(new_dict)
    return pd.DataFrame(new_data)

# convert json to csv
with open(root+f_name, 'r') as json_file:
    json_string = json.load(json_file)
event_df = parse_json_string(json_string)
event_df.to_csv(root+f_name[:-5]+'.csv')
```
